# 1) Área de un triángulo (base y altura)

```python
base = float(input("Ingrese la base del triángulo: "))
altura = float(input("Ingrese la altura del triángulo: "))

area = (base * altura) / 2
print("El área del triángulo es:", area)
```

# 4) Perímetro y área de un círculo (radio)

```python
import math

radio = float(input("Ingrese el radio del círculo: "))

perimetro = 2 * math.pi * radio
print("El perímetro del círculo es:", perimetro)

area = math.pi * radio * radio
print("El área del círculo es:", area)
```

# 5) Resolver ecuación cuadrática (ax² + bx + c = 0)

```python
import math

a = float(input("Ingrese a: "))
b = float(input("Ingrese b: "))
c = float(input("Ingrese c: "))

d = b**2 - 4*a*c

if d > 0:
    x1 = (-b + math.sqrt(d)) / (2*a)
    x2 = (-b - math.sqrt(d)) / (2*a)
    print("Dos soluciones reales:")
    print("x1 =", x1)
    print("x2 =", x2)
elif d == 0:
    x = -b / (2*a)
    print("Una solución real:")
    print("x =", x)
else:
    print("No tiene soluciones reales.")
```

# 10) Determinar tipo de triángulo (equilátero, isósceles o escaleno)

```python
a = float(input("Ingrese el lado a: "))
b = float(input("Ingrese el lado b: "))
c = float(input("Ingrese el lado c: "))

if a == b and b == c:
    print("El triángulo es equilátero.")

elif a == b or a == c or b == c:
    print("El triángulo es isósceles.")

else:
    print("El triángulo es escaleno.")
```

# 12) Calculadora simple (+, -, *, /)

# 13) Calificación en letras (0–100)

```python
nota = int(input("Ingrese la nota (0 a 100): "))

if nota < 0 or nota > 100:
    print("Nota inválida.")
elif nota >= 90:
    print("Calificación: A")
elif nota >= 80:
    print("Calificación: B")
elif nota >= 70:
    print("Calificación: C")
elif nota >= 60:
    print("Calificación: D")
else:
    print("Calificación: F")
```

# 14) Día de la semana según número (1–7)

```python
n = int(input("Ingrese un número del 1 al 7: "))

if n == 1:
    print("Lunes")

elif n == 2:
    print("Martes")

elif n == 3:
    print("Miércoles")

elif n == 4:
    print("Jueves")

elif n == 5:
    print("Viernes")

elif n == 6:
    print("Sábado")

elif n == 7:
    print("Domingo")

else:
    print("Número inválido. Debe ser del 1 al 7.")
```

# 15) Validar si forman triángulo y decir el tipo

```python
a = float(input("Ingrese el lado 1: "))
b = float(input("Ingrese el lado 2: "))
c = float(input("Ingrese el lado 3: "))

if a + b <= c or a + c <= b or b + c <= a:
    print("No se puede formar un triángulo con esos lados.")

else:

    if a == b and b == c:
        print("El triángulo es equilátero.")

    elif a == b or a == c or b == c:
        print("El triángulo es isósceles.")

    else:
        print("El triángulo es escaleno.")
```
